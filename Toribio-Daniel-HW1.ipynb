### Model formulation

#### Decision variables

We have to define indices to identify the decision variables:

\begin{equation}
i \in \text{I}=\{1,2,\dots,15\}: \text{Set of products.}
\end{equation}

\begin{equation}
j \in \text{J}=\{\text{Jan},\text{Feb},\dots,\text{Dec}\}: \text{Set of months.}
\end{equation}

\begin{equation}
k \in \text{K}=\{\text{Grinder},\text{Drill},\dots,\text{Stamping}\}: \text{Set of machines.}
\end{equation}

Firstly we have to identify what are the decision variables. In this problem we can make decisions in the number of products that will be produced each month, the number of products that will be sold each month, and the number of products that we are going to store in the warehouse each month.

\begin{equation}
\text{make}_{i,j} \in \mathbb{R} : \text{Number of units of product i to make at month j.} 
\end{equation}

\begin{equation}
\text{sell}_{i,j} \in \mathbb{R} : \text{Number of units of product i to sell at month j.}
\end{equation}

\begin{equation}
\text{store}_{i,j} \in \mathbb{R} : \text{Number of units of product i to store at month j.}
\end{equation}

#### Objective Function

Once we have defined our decision variables we can define an objective function, in this case we want to maximize the total profit of the year:

\begin{equation}
\text{Max} : \sum_{i}\sum_{j}
(\text{profit}_i \times \text{sell}_{i,j} - \text{0.2} \times \text{store}_{i,j})
\end{equation}

#### Constraints

We have different constraints:

- For the first month, January, the number of units manufactured must be equal to the number of sold units plus the number of stored units. This coinstraint is necessary to avoid sending units that are not produced yet.

\begin{equation}
\text{make}_{i,j} = \text{sell}_{i,j} + \text{store}_{i,j} \quad \forall i \in \text{I} \quad j = \text{Jan}
\end{equation}


- For the rest of the months we also need to add a constraint about the balance to avoid sending units which do not exist. This can be modeled as the number of stored units of the previous year plus the number of manufactured units must be equal to the number of sold units plus the number of stored units each month. For this constraint we do not consider January.

\begin{equation}
\text{store}_{i,j-1} + \text{make}_{i,j} = \text{sell}_{i,j} + \text{store}_{i,j} \quad \forall i \in \text{I} \quad \forall j \in \text{J} \setminus \{\text{Jan}\}
\end{equation}


- We have to take care about the number of products, of each type, the factory can produce. I will refer to that limitation for product i on month j as make_limit_i,j.

\begin{equation}
\text{make_limit}_{i,j} : \text{Maximum number of products i that can be made on month j.}
\end{equation}

\begin{equation}
\text{make}_{i,j} \leq \text{make_limit}_{i,j} \quad \forall i \in \text{I} \quad \forall j \in \text{J} 
\end{equation}


- We have to consider the market limitations in the number of products that we are going to sell. For that I will refer to the limit for selling products on month j as sell_limit_j.

\begin{equation}
\text{sell_limit}_{j} : \text{Maximum amount of products that can be sold on month j.}
\end{equation}

\begin{equation}
\sum_{i}\text{sell}_{i,j} \leq \text{sell_limit}_{j} \quad \forall j \in \text{J} 
\end{equation}



- We need to satisfy the minimum demand of each product each month. For that I will define the minimum demand of product i at month j as demand_i,j.

\begin{equation}
\text{demand}_{i,j} : \text{Minimum demand of product i at month j.}
\end{equation}


\begin{equation}
\text{sell}_{i,j} \geq \text{demand}_{i,j} \quad \forall i \in \text{I} \quad \forall j \in \text{J} 
\end{equation}

- As we have a certain number of hours per month (304 hours) and each product needs a certain time to be made, the time required to manufacture the total number of products must be smaller or equal to the number of total working hours. To define in a compact way this constraint I will refer to the time to make product i with machine k as time_i,k.

\begin{equation}
\text{time}_{i,k} : \text{Time required to make product i with machine k.}
\end{equation}

\begin{equation}
\sum_{k}\text{time}_{i,k} \times \text{make}_{i,j} \leq \text{304} \quad \forall i \in \text{I} \quad \forall j \in \text{J} 
\end{equation}

- The warehouse can not store more than 800 products:

\begin{equation}
\sum_{i}\text{store}_{i,j} \leq \text{3000} \quad \forall j \in \text{J} 
\end{equation}

### Model implementation

Once the model is formulated we just need to implement it in pyomo.First, we have to create the model, set the indices and then add to the model the decision variables, the objective function and the constraints.

```python
model = ConcreteModel("Linear model")
model.I = Set(initialize = Products, doc = "Products")
model.J = Set(initialize = Months, doc = 'Months')
model.K = Set(initialize = Machines, doc= "Machines")

model.Profit = Param(model.I, initialize = profit, doc="Profit")
model.Time = Param(model.I, model.K, initialize = time, doc = "Time")
model.Make_limit = Param(model.J, model.I, initialize = make_limit, doc="Make limit")
model.Sell_limit = Param(model.J, initialize = sell_limit, doc = "Sell limit")
model.Demand = Param(model.J, model.I, initialize = demand, doc = "Demand")

model.make = Var(model.I, model.J, doc='Units of product made', within=NonNegativeReals)
model.sell = Var(model.I, model.J, doc='Units of product sold', within=NonNegativeReals)
model.store = Var(model.I, model.J, doc='Units of product stored', within=NonNegativeReals)

def objective_rule(model):
    return sum(model.Profit[i]*model.sell[i,j]-0.2*model.store[i,j] for i in model.I for j in model.J)

model.objective = Objective(rule=objective_rule, sense=maximize, doc='Total profit')

def C1(model,i):
    return model.make[i,"Jan"] == model.sell[i,"Jan"] + model.store[i,"Jan"]

model.C1 = Constraint(model.I, rule=C1, doc='Initial balance')

def C2(model,i,j):
    return model.store[i,Months[Months.index(j)-1]] + model.make[i,j] == model.sell[i,j] + model.store[i,j]

model.C2 = Constraint(model.I, model.J, rule=C2, doc='General balance')

def C3(model,i,j):
    return model.make[i,j] <= model.Make_limit[j,i]

model.C3 = Constraint(model.I,model.J, rule=C3, doc='Make limit')

def C4(model,j):
    return sum(model.sell[i,j] for i in model.I) <= model.Sell_limit[j]

model.C4 = Constraint(model.J, rule=C4, doc='Sell limit')

def C5(model,i,j):
    return model.sell[i,j]  >= model.Demand[j,i]

model.C5 = Constraint(model.I,model.J, rule=C5, doc='Demand')

def C6(model,i,j):
    return sum(model.Time[i,k] * model.make[i,j] for k in model.K) <=304

model.C6 = Constraint(model.I,model.J, rule=C6, doc='Time limit')

def C7(model,j):
    return sum(model.store[i,j] for i in model.I) <= 3000

model.C7 = Constraint(model.J, rule=C7, doc='Store limit')
```

Once we have implemented the model, we have to solve it. I will use gurobi solver as it is faster than other solvers.

```python
Solver = SolverFactory('gurobi')
solution = Solver.solve(model)
```

### Adding logical and conditional constraints

The factory told us that the machines used to manufacture the products needs a month for maintenance. The employees who makes these maintenances said that they can only work with one machine per month. The factory wants to know in which month they should repair each machine in order to maxime the profit.



#### Decision variables

Now we are also able to make decisions in the machine that is going to be at maintenance in a certain month. For that I will define a binary variable where 0 means that the machine k is not going to be at maintenance in the month j, and 1 means that it will be in maintenance.

\begin{equation}
\text{maintenance}_{k,j} \in \text{\{0,1\}} : \text{If machine k is in maintenance in month j.}
\end{equation}

If a machine used to make certain products is in maintenance, we will not be able to make those products. I will create another binary variable to define if a product i can be manufactured in month j.

\begin{equation}
\text{do}_{i,j} \in \text{\{0,1\}} : \text{If product i can be manufactured in month j.}
\end{equation}



#### Constraints

After including the binary variable maintenance we have to change some of the constraints and add new ones:

- First, we have to ensure that each machine has been in maintenance for one month.

\begin{equation}
\sum_{j}\text{maintenance}_{k,j} = \text{1}  \quad \forall k \in \text{K} 
\end{equation}

- Now we have to ensure that each month at most one machine is in maintenance.

\begin{equation}
\sum_{k}\text{maintenance}_{k,j} \leq \text{1}  \quad \forall j \in \text{J} 
\end{equation}

- If a machine is in maintenance, certain products can not be manufactured. Note that we have to particularize for each product as depending on the type we have to use different machines. For example for P1 only the grinder, drill, and the stamping machine are needed. This means that if the milling machine is in maintenance it will not affect the manufacturing of P1.

\begin{equation}
\text{do}_{i,j} \leq 1 - \sum_{k \in \text{\{grinder, drill, stamping\}}}\text{maintenance}_{k,j}  \quad i = \text{P1} \quad \forall j \in \text{J} \quad
\end{equation}

This constraint has to be repeated for each product taking the sum of only the machines which are needed for that product.


- The constraint about time now changes as if we have a certain machine in maintenance in a certain month, products that need that machine can not be produced that month.

\begin{equation}
\sum_{k}\text{time}_{i,k} \times \text{make}_{i,j} \leq \text{304} \times \text{do}_{i,j} \quad \forall i \in \text{I} \quad \forall j \in \text{J} 
\end{equation}


```python
model = ConcreteModel("Binary model")
model.I = Set(initialize = Products, doc = "Products")
model.J = Set(initialize = Months, doc = 'Months')
model.K = Set(initialize = Machines, doc= "Machines")

model.Profit = Param(model.I, initialize = profit, doc="Profit")
model.Time = Param(model.I, model.K, initialize = time, doc = "Time")
model.Make_limit = Param(model.J, model.I, initialize = make_limit, doc="Make limit")
model.Sell_limit = Param(model.J, initialize = sell_limit, doc = "Sell limit")
model.Demand = Param(model.J, model.I, initialize = demand, doc = "Demand")

model.make = Var(model.I, model.J, doc='Units of product made', within=NonNegativeReals)
model.sell = Var(model.I, model.J, doc='Units of product sold', within=NonNegativeReals)
model.store = Var(model.I, model.J, doc='Units of product stored', within=NonNegativeReals)
model.maintenance = Var(model.K, model.J, doc='If is in maintenance', within=Binary)
model.do = Var(model.I, model.J,  doc='If a product can be manufactured', within=Binary)

def objective_rule(model):
    return sum(model.Profit[i]*model.sell[i,j]-0.2*model.store[i,j] for i in model.I for j in model.J)

model.objective = Objective(rule=objective_rule, sense=maximize, doc='Total profit')

def C1(model,i):
    return model.make[i,"Jan"] == model.sell[i,"Jan"] + model.store[i,"Jan"]

model.C1 = Constraint(model.I, rule=C1, doc='Initial balance')

def C2(model,i,j):
    return model.store[i,Months[Months.index(j)-1]] + model.make[i,j] == model.sell[i,j] + model.store[i,j]

model.C2 = Constraint(model.I, model.J, rule=C2, doc='General balance')

def C3(model,i,j):
    return model.make[i,j] <= model.Make_limit[j,i]

model.C3 = Constraint(model.I,model.J, rule=C3, doc='Make limit')

def C4(model,j):
    return sum(model.sell[i,j] for i in model.I) <= model.Sell_limit[j]

model.C4 = Constraint(model.J, rule=C4, doc='Sell limit')

def C5(model,i,j):
    return model.sell[i,j]  >= model.Demand[j,i]

model.C5 = Constraint(model.I,model.J, rule=C5, doc='Demand')


def C6(model,i,j):
    return sum(model.Time[i,k] * model.make[i,j] for k in model.K) <= (304 * model.do[i,j] )

model.C6 = Constraint(model.I,model.J, rule=C6, doc='Time limit')


def C7(model,k):
    return sum(model.maintenance[k,j] for j in model.J) == 1

model.C7 = Constraint(model.K, rule=C7, doc='Each machine is in maintenance for one month')

def C8(model,j):
    return sum(model.maintenance[k,j] for k in model.K) <= 1

model.C8 = Constraint(model.J, rule=C8, doc='Only one machine in maintenance in a month')

def C9(model,j):
    return sum(model.store[i,j] for i in model.I) <= 3000

model.C9 = Constraint(model.J, rule=C9, doc='Store limit')

def CP1(model,j):
    return model.do["P1",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Drill","Stamping"])

model.CP1 = Constraint(model.J, rule=CP1, doc='Logical constraint')

def CP2(model,j):
    return model.do["P2",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Lathe","Stamping"])

model.CP2 = Constraint(model.J, rule=CP2, doc='Logical constraint')

def CP3(model,j):
    return model.do["P3",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Milling","Stamping"])

model.CP3 = Constraint(model.J, rule=CP3, doc='Logical constraint')

def CP4(model,j):
    return model.do["P4",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Lathe","Stamping"])

model.CP4 = Constraint(model.J, rule=CP4, doc='Logical constraint')

def CP5(model,j):
    return model.do["P5",j] <= 1 - sum(model.maintenance[k,j] for k in ["Lathe","Milling","Waterjet"])

model.CP5 = Constraint(model.J, rule=CP5, doc='Logical constraint')

def CP6(model,j):
    return model.do["P6",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Lathe","Waterjet","Presses","Stamping"])

model.CP6 = Constraint(model.J, rule=CP6, doc='Logical constraint')

def CP7(model,j):
    return model.do["P7",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Lathe","Milling"])

model.CP7 = Constraint(model.J, rule=CP7, doc='Logical constraint')

def CP8(model,j):
    return model.do["P8",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Milling","Presses","Stamping"])

model.CP8 = Constraint(model.J, rule=CP8, doc='Logical constraint')

def CP9(model,j):
    return model.do["P9",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Lathe","Waterjet"])

model.CP9 = Constraint(model.J, rule=CP9, doc='Logical constraint')

def CP10(model,j):
    return model.do["P10",j] <= 1 - sum(model.maintenance[k,j] for k in ["Waterjet","Presses"])

model.CP10 = Constraint(model.J, rule=CP10, doc='Logical constraint')

def CP11(model,j):
    return model.do["P11",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Drill","Milling","Presses","Stamping"])

model.CP11 = Constraint(model.J, rule=CP11, doc='Logical constraint')

def CP12(model,j):
    return model.do["P12",j] <= 1 - sum(model.maintenance[k,j] for k in ["Drill","Presses"])

model.CP12 = Constraint(model.J, rule=CP12, doc='Logical constraint')

def CP13(model,j):
    return model.do["P13",j] <= 1 - sum(model.maintenance[k,j] for k in ["Milling","Stamping"])

model.CP13 = Constraint(model.J, rule=CP13, doc='Logical constraint')

def CP14(model,j):
    return model.do["P14",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Drill","Stamping"])

model.CP14 = Constraint(model.J, rule=CP14, doc='Logical constraint')

def CP15(model,j):
    return model.do["P15",j] <= 1 - sum(model.maintenance[k,j] for k in ["Grinder","Stamping"])

model.CP15 = Constraint(model.J, rule=CP15, doc='Logical constraint')
```
